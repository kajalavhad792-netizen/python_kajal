<h3>For loop</h3>

A for loop is used when you know how many times you want to repeat something or when you want to iterate through a sequence like a list, string, tuple, or range.

#Syntax

for variable in sequence:

#code


```python
# Q1.Print your name 10 times.

for i in range(10):
    print("kajal")
```

```python
# Q2. Print numbers from 1 to 100. 

for i in range(1,101):
    print(i)
```

```python
# Q3.Print the square of numbers from 1 to 10.
for i in range(1,11):
    print(i * i)
    
```

```python
# Q4. Print the cube of numbers from 1 to 10. 


for i in range(1,11):
    print(i * i * i)
```

```python
# Q5. Find the sum of numbers from 1 to 100.

sum = 0
for i in range(1,101):
    sum = sum + i
    print(i,end=" ")
   
```

```python
#Q6.Count how many numbers are divisible by 3 between 1 and 50. 

count = 0
for i in range(1,51):
    if i % 3 ==0:
         count = count + 1
print("count =", count)
     
    
```

```python
# Q7. Print the first 10 natural numbers in reverse order.

for i in range(10, 0, -1):
    print(i)
```

```python
# Q8. Print all vowels from the string: 
#text = "FutureBright"

text = "FutureBright"
for char in text:
    if char in "aeiouAEIOU":
        print(char)
```

```python
#Q10. Count how many numbers are greater than 50. 
 #marks = [45, 78, 90, 34, 88, 51]

marks = [45, 78, 90, 34, 88, 51]

count = 0

for mark in marks:
    if mark > 50:
        count = count + 1

print("Count =", count)
```
