```python
# Find the maximum, minimum, total, and average of a tuple containing student marks.

marks = (85, 92, 78, 88, 95)

maximum = max(marks)
minimum = min(marks)
total = sum(marks)
average = total / len(marks)

print("Maximum Marks:", maximum)
print("Minimum Marks:", minimum)
print("Total Marks:", total)
print("Average Marks:", average)
```

variables.

```python
# Write a function that returns a student's total marks and average. Unpack both values into separate variables

def student_marks(marks):
    total = sum(marks)
    average = total / len(marks)
    return total, average
marks = [85, 90, 78, 88, 95]
total_marks, average_marks = student_marks(marks)
print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
```

```python
# Given a tuple (101, "Rahul", "Developer", 65000), unpack it into meaningful variable names.

employee = (101, "Rahul", "Developer", 65000)
emp_id, name, designation, salary = employee
print("Employee ID :", emp_id)
print("Name        :", name)
print("Designation :", designation)
print("Salary      :", salary)
```

```python
# Use star unpacking to separate the first element from the remaining elements.

numbers = (10, 20, 30, 40, 50)
first, *remaining = numbers
print("First Element :", first)
print("Remaining     :", remaining)
```

Mini Project 
Employee Salary Report 

employees = ( 

("Rahul", 50000), 

("Amit", 65000), 

("Sneha", 72000), 

("Pooja", 58000) 

) 

Perform the following:

1. Print each employee's name and salary.

2. Create a tuple containing only the salaries.


3. Find:

   
o Highest salary  

o Lowest salary  

o Total salary expenditure 

o Average salary



```python
# 1. Print each employee's name and salary


employees = (
    ("Rahul", 50000),
    ("Amit", 65000),
    ("Sneha", 72000),
    ("Pooja", 58000)
)



print("Employee Salary Details:")
for name, salary in employees:
    print(name, ":", salary)
```

```python
# 2. Create a tuple containing only the salaries

salaries = tuple(salary for name, salary in employees)
print("\nSalaries Tuple:", salaries)
```

```python
#  3. Salary Report

highest_salary = max(salaries)
lowest_salary = min(salaries)
total_salary = sum(salaries)
average_salary = total_salary / len(salaries)

print("\nSalary Report")
print("Highest Salary:", highest_salary)
print("Lowest Salary:", lowest_salary)
print("Total Salary Expenditure:", total_salary)
print("Average Salary:", average_salary)
```
