<h3>Truthy</h3>

A truthy value is any value that Python treats as True when used in a condition (if, while, etc.).

Non-zero number → Truthy ✅

Non-empty string → Truthy ✅

Non-empty list/tuple/set/dictionary → Truthy ✅

<h3>Falsy</h3>

A falsy value is any value that Python treats as False when used in a condition (if, while, etc.)

Zero value → Falsy ❌

Empty string → Falsy ❌

Empty list/tuple/dictionary/set → Falsy ❌

False and None → Falsy ❌

<h3>Match-Case</h3>

The match-case statement is used to compare a value against multiple cases

```python
#Assignment 2: Traffic Signal using match-case
#Create a traffic signal program.
#Input:
#Signal Color
#Cases:
#Red → Stop
#Yellow → Get Ready
#Green → Go
#Any other value → Invalid Signal

#Use the match-case statement.


signal = (input("enter a signal:"))

match signal:
    case "red":
        print("stop")
    case "yellow":
        print("get ready")
    case "green":
        print("go")
    case _:
        print("invalid signal")

    
         
```

```python
#  Calculator using match-case

#Create a calculator.

#Menu:
#Addition
#Subtraction
#Multiplication
#Division
#Task:
#Ask the user to choose an operation.
#Take two numbers as input.
#Perform the selected operation using match-case



number = int(input("enter a number (1-10):"))

num1 = int  (input("Enter first number: "))
num2 = int  (input("Enter second number: "))

match number:
    case 1:
        print("addition = ", num1 + num2)
    case 2:
        print("substraction =",num1 - num2)
    case 3:
        print("multiplication = ",num1 * num2)
    case 4:
        print("division = ",num1 / num2)
    case _:
         print("invalid option")
```

```python
# Assignment 7: Banking Deposit System

#Create a banking deposit program.

#Input:
#Account Active (True/False)
#KYC Completed (True/False)
#Deposit Amount
#Rules:
#If the account is inactive, reject the deposit.
#If KYC is incomplete, ask the customer to complete KYC.
#If the deposit amount is less than or equal to zero, display "Invalid Deposit Amount".
#If the deposit amount is greater than ₹50,000, display "PAN Verification Required".
#If the deposit amount is greater than ₹2,00,000, display "Manager Approval Required".
#Otherwise, complete the deposit successfully.
45

account_active = input("Is the account active? (True/False): ") == "True"
kyc_completed = input("Is KYC completed? (True/False): ") == "True"
deposit_amount = float(input("Enter deposit amount: "))

if not account_active:
    print("Deposit Rejected. Account is Inactive.")

elif not kyc_completed:
    print("Please Complete KYC.")

elif deposit_amount <= 0:
    print("Invalid Deposit Amount.")

elif deposit_amount > 200000:
    print("Manager Approval Required.")

elif deposit_amount > 50000:
    print("PAN Verification Required.")

else:
    print("Deposit Successful.")
```

```python
# Assignment 11: Mini Banking Application

#Create a banking application with the following menu:

#1. Deposit
#2. Withdraw
#3. Balance Enquiry
#4. Exit
#Requirements:
#Use match-case to handle menu options.
#Use combined conditional statements for transaction validation.
#Use Truthy/Falsy values wherever applicable.
#Display meaningful messages for every operation.
#Handle invalid menu choices using the default (case _) option.


balance = 1000



choice = int(input("Enter your choice: "))

match choice:

        case 1:
            amount = int(input("Enter deposit amount: "))

            if amount > 0:
                balance += amount
                print("Deposit Successful!")
                print("Current Balance:", balance)
            else:
                print("Invalid deposit amount.")

        case 2:
            amount = int(input("Enter withdrawal amount: "))

            if amount > 0 and balance >= amount:
                balance -= amount
                print("Withdrawal Successful!")
                print("Current Balance:", balance)
            elif amount <= 0:
                print("Invalid withdrawal amount.")
            else:
                print("Insufficient Balance.")

        case 3:
            if balance:
                print("Current Balance:", balance)
            else:
                print("Your account balance is zero.")

        case 4:
            print("Thank you for banking with us!")
            

        case _:
            print("Invalid Choice! Please select a valid option (1-4).")


    
```
