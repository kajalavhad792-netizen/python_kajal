```python
# Print numbers from 1 to 10.

for i in range(1,11):
    print(i,end=" ")
```

```python
# Print numbers from 10 to 1.

for i in range(11,-1):
    print(i,end=" ")
```

```python
# Print all even numbers from 1 to 20.

for i in range(1,21):
    if i % 2==0:
        print(i,end=" ")
```

```python
# Print all odd numbers from 1 to 20.

for i in range(1,21):
    if i % 2 != 0:
        print(i,end=" ")
```
