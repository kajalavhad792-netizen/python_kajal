<h3>Conditional statements</h3

A conditional statement is used to make decisions in a program. It executes a block of code only if a specified condition is True.


1.  Conditional statements help a program make decisions.

2.  A condition always returns True or False.

3.  Python uses Boolean expressions to check conditions.

   

<h4>Types of Conditional statements</h4>

1. If
2. If...Else
3. If...Elif...Else
4. Nested If

<h4>1.If</h4>

The if statement execute the code only when the condition is True




<h4>2. If...Else</h4>

The if...else statement allows a program to choose one of two possible paths based on a condition

what is if..else

The if..else statement gives the program two choices

1. If the condition is true --> execute the if block

2. Otherwise --> execute the else block

* Only one block is executed

✓ if-else is used for two-way decision making. 

✓The if block executes when the condition is True. 

✓The else block executes when the condition is False. 

✓Only one block executes during program execution. 

✓else does not have any condition. 

✓ Indentation is mandatory for both blocks. 

✓ if-else improves decision-making in programs. 

✓ It is widely used in real-world applications like login 
systems and banking.

<h4>3. If...Elif</h4>

The if...elif statement is used to check multiple conditions.

If the first condition is False, Python checks the next elif condition. The first condition that evaluates to True is executed.

* How if-elif-else statement works?
  
 Python checks conditions from top to bottom.
 
•If the first condition is True, its block executes. 

•Python skips all remaining elif and else blocks. 

•If the first condition is False, Python checks the next elif.

•This continues until a condition becomes True. 

•If no condition is True, the else block executes.

<h4>4. Nested if</h4>

A nested if statement is an if statement inside another if statement. It is used when a second condition should be checked only if the first condition is True.

•A Nested if is an if statement inside another if. 

•The inner if executes only if the outer if condition is True. 

•Nested if is used when one condition depends on another. 

•Multiple levels of nesting are possible.

•Proper indentation is very important. 

•Too many nested if statements make code difficult to read. 

•Nested if is commonly used in authentication and validation systems. 

•Developers should avoid unnecessary nesting by writing clean and simple logic

<h4>4. Nested if </h4>

```python
# Check if a student is eligible for an exam based on attendance, then check if fees are paid.

attendance = int(input("Enter attendance percentage: "))
fees_paid = input("Have you paid the fees? (yes/no): ")

if attendance >= 75:
    if fees_paid == "yes":
        print("Eligible for Exam")
    else:
        print("Fees not paid")
else:
    print("Attendance is less than 75%")
```

```python
# using AND

attendance = int(input("Enter attendance percentage: "))
fees_paid = input("Have you paid the fees? (yes/no): ")

if attendance >= 75 and fees_paid == "yes":
    print("Eligible for Exam")
else:
    print("Not Eligible for Exam")
```

```python
# Create a login system that checks username and then password.

username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin":
    if password == "1234":
        print("Login Successful")
    else:
        print("Incorrect Password")
else:
    print("Invalid Username")
```

```python
# using AND

username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Username or Password")
```

```python
# Create a college admission system that checks marks and then document verification.

marks = int(input("Enter marks: "))
documents = input("Documents verified? (yes/no): ")

if marks >= 60:
    if documents == "yes":
        print("Admission Confirmed")
    else:
        print("Documents Pending")
else:
    print("Marks are insufficient")
```

# Compare solutions using Nested if and logical operators (and) and identify which is more readable.


| Nested if                                           | Logical Operator (`and`)                          |
| --------------------------------------------------- | ------------------------------------------------- |
| Checks one condition, then another inside it.       | Checks multiple conditions in a single statement. |
| More lines of code.                                 | Less code.                                        |
| Useful when each condition has a different message. | Useful when only the final result matters.        |
| Easier to handle multiple outcomes.                 | Easier to read for simple conditions.             |


<h4>Which is More Readable?</h4>

For simple conditions, using the and operator is more readable because it is shorter and easier to understand.

For complex problems, where each condition requires a different action or message, Nested if is more suitable.
