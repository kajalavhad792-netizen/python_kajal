<h3>Assignment I : String Formatting</h3>

```python
#Display your biodata using string concatenation

name = "Kajal"
age = 21
city = "Nashik"

print("Name: " + name)
print("Age: " + str(age))
print("City: " + city)
```

```python
#Display the same biodata using % formatting.

name = "Kajal"
age = 21
city = "Nashik"

print("Name: %s " % name)
print("Age: %d " % (age))
print("City: %s " % city)
```

```python
# display it using str.format()

name = "Kajal"
age = 21
city = "Nashik"

print("Name: {}".format(name))
print("Age: {}".format(age))
print("City: {}".format(city))
```

```python
# Display it using f-Strings.

name = "Kajal"
age = 21
city = "Nashik"

print(f"Name: {name}")
print(f"Age: {age}")
print(f"City: {city}")
```

```python
# Print the sum of two numbers inside an f-String.

num1 = 20
num2 = 30

print(f"Sum = {num1 + num2}")
```

```python
# Create a formatted shopping bill using variables.

item = "Notebook"
price = 80
quantity = 3
total = price * quantity

print("----- Shopping Bill -----")
print(f"Item     : {item}")
print(f"Price    : ₹{price}")
print(f"Quantity : {quantity}")
print(f"Total    : ₹{total}")
```
