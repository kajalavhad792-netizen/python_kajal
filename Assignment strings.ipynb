<h3> Assignment 1 :Basic string operation</h3>

```python
# find the lenght of the string

name = "kajal avhad"
print(len(name))
```

```python
# print the first and last character

name = "kajal avhad"
print("first character:", name [0])
print("second character:", name [-1])
```

```python
# print the third character

name = "kajal"
print(name [2])
```

```python
# print the last three characters

name = "kajal"
print(name [-3:])
```

```python
# reverse the string using slicing

name = "kajal"
print(name [::-1])
```

```python
# print every second character

name = "kajal avhad"
print(name [::2])
```

```python
# print the string in reverse order using negative indexing

name = "kajal avhad"
print(name [::-1])
```

```python
# create a string of containig your fullname and print it

name = "Kajal Avhad"
print(name)
```

<h3>Assignment B :Indexing & Slicing </h3>

```python
# extract the first name from your full name

name = "kajal avhad"
print(name [:5])
```

```python
# extract the last name

name = "kajal avhad"
print(name [5:])
```

```python
# extract the first five characters

name = "kajal avhad"
print(name [:5])
```

```python
# print all characters except first and last

name = "kajal avhad"
print(name[1:-1])
```
