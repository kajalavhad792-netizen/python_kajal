<h3>Assignment 3: while Loop</h3>

```python
# Q1. Print numbers from 1 to 20.

i = 1
while i <= 20:
    print(i)
    i += 1
```

```python
# Q2. Print numbers from 20 to 1.

i = 20
while i >= 1:
    print(i)
    i -= 1
```

```python
# Q3. Print all even numbers between 1 and 50.

i = 2
while i <= 50:
    print(i)
    i = i + 2
```

```python
# Q4. Print the multiplication table of 9. 

i = 1
while i <= 10:
    print(9 * i)
    i += 1
```

```python
# Q5. Print numbers until they reach 100, increasing by 10 each time. 

i = 10
while i <= 100:
    print(i)
    i += 10
```

```python
# Q6. Display a countdown from 10 to 1, then print "Happy New Year!".

i = 10

while i >= 1:
    print(i)
    i = i - 1

print("Happy New Year!")
```

```python
# Q7. Simulate an ATM balance starting at ₹5000, withdrawing ₹500 until the balance becomes 0.

balance = 5000

while balance > 0:
    print("Current Balance: ₹", balance)
    balance -= 500

print("Balance is ₹0")
print("Insufficient Balance!")
```

```python
# Q8. Ask the user to enter the password repeatedly until they enter: 
#python123 

password = ""

while password != "python123":
    password = input("Enter password: ")

print("Access Granted!")
```
