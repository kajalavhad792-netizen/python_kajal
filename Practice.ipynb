```python
#Calculate the electricity bill based on units:
'''
0–100 → ₹2/unit
101–200 → ₹3/unit
Above 200 → ₹5/unit
'''
bill = int(input('enter a bill:'))
if bill >= 0 and bill <= 100:
    print('₹2/unit',bill * 2)
elif bill >= 101 and bill <= 200:
    print('₹3/unit',bill * 3)
elif bill > 200:
    print('₹5/unit',bill * 5)
else:
    print('invalid bill')
```

```python
# Write a menu-driven program:
'''
1 → Addition
2 → Subtraction
3 → Multiplication
4 → Division
'''

choice = int(input("Enter your choice (1-4): "))

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

if choice == 1:
    print("Addition =", num1 + num2)
elif choice == 2:
    print("Subtraction =", num1 - num2)
elif choice == 3:
    print("Multiplication =", num1 * num2)
elif choice == 4:
    print("Division =", num1 / num2)
else:
    print("Invalid Choice")
```
