<h3> Loops </h3>

Loops are used to execute a block of code repeatedly until a condition is met or for every item in a sequence.

<h4>Types of loops</h4>

<h4>1.For loop</h4>

A for loop is used when you know how many times you want to repeat something or when you want to iterate through a sequence like a list, string, tuple, or range.

<h5>#Syntax</h5>

for variable in sequence:

    # code

<h4>2.While loop</h4>

A while loop runs as long as a condition is True.

<h5>#Syntax</h5>

while condition:

    # code

<h3>Assignments</h3>

<h4>Assignment 1: range() Function</h4>

```python
#Q1. 
#Print numbers from 1 to 20.

for i in range(1,21):
    print(i)
```

```python
# Q2. 
#Print numbers from 20 to 1. 

for i in range(21,0,-1):
    print(i)
```

```python
# Q3. 
#Print all even numbers between 1 and 50.

for i in range(1,51):
    if i % 2 ==0:
      print(i)
```

```python
# Q4. 
#Print all odd numbers between 1 and 50.

for i in range(1,51):
    if i % 2 != 0:
        print(i)
```

```python
# Q5. 
#Print the multiplication table of 7 using range().

for i in range(1, 11):
    print( 7 * i)
```

```python
# Q6. 
#Print numbers from 100 to 10 with a gap of 10. 

for i in range(100,9,-10):
    print(i)
```

```python
# Q7. 
#Print all multiples of 5 between 1 and 100.

for i in range(5,101,5):
    print(i)
```
