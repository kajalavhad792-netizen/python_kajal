<h3>Escape Characters</h3>

An escape character is a special character used inside a string to represent characters that cannot be typed directly or have a special meaning.

<h4>Types of Escape Characters</h4>

1. \n = New line (move to the next line)

2. \t = Tab (adds a horizontal indent)

3. \\ = Backslash (prints a literal)

4. \" = Double qoute (allows double qoutes inside the double qouted string)

5. \' = Single qoute (allows single qoutes inside single qouted string)


<h4>Examples</h4>

<h4>1.\n   New line</h4>

```python
a = ("hallo\nworld")
print(a)
```

```python
b = ("kajal\n21")
print(b)
```

```python
c = ("Name: Kajal\nAge: 22")
print(c)
```

```python
lang = ("python\nprogramming")
print(lang)
```

```python
line = ("line1\nline2\nline3")
print(line)
```

<h4>2. \t   Tab</h4>

```python
lang = ("hallo\tworld")
print(lang)
```

```python
a = ("name:kajal\tage:21")
print(a)
```

```python
fruit = ("apple\tbanana\tmango")
print(fruit)
```

<h3>3.\\  Backslash</h3>

```python
a =("C:\\Users")
print(a)
```

```python
b = ("This is a backslash: \\")
print(b)
```

```python
lang = ("D:\\Python\\Programs")
print(lang)
```

<h4>4. \"   Double qoute </h4>

```python
a = ("He said \"Hello\"")
print(a)
```

```python
b = ("my name is \"kajal\"")
print(b)
```

```python
qoute = ("The teacher said, \"Practice every day.\"")
print(qoute)
```

<h4>5.\'   Single qoute</h4>

```python
a = ('it\'s me')
print(a)
```

```python
lang = ('it\'s python programming')
print(lang)
```

```python
b = ('I\'m Kajal')
print(b)
```
