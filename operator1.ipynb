<h3>1. Assignment Operator</h3>

 Is used to assign the value to the variable
 

<h4>Types of Assignment Operator</h4>

1. (=) = Assign a value
2. (+=) = Add and Assign
3. (-=) = Subtract and Assign
4. (*=) = Multiply and Assign
5. (/=) = Divide and Assign
6. (%=) = Modulus and Assign
7. (//=) = Floor division and Assign
8. (**=) = Power and Assign

1.( = )  <h4>Assign a value</h4>

x = 10

 <H4>2.Add and Assign  (+=)</H4>
 

```python
x = 10

x = x + 5

print(x)
```

<h4>3. Subtract and Assign (-=)</h4>

```python
x = 10

x = x - 5

print(x)
```

<h4>4.Multiply and Assign (*=)</h4>

```python
x = 10

x = x * 5

print(x)
```

<h4>5. Divide and Assign (/=)</h4>

<h4>6. Modulus and Assign (%=)</h4>

```python
x = 10

x = x % 5

print(x)
```

<h4>7. Floor division and Assign (//=)</h4>

```python
x = 10

x = x // 5

print(x)
```

<h3>2. Membership Operator</h3>

Used to check whether the value exist in sequence such as list,tuple,set,dictionary,string

<h4> Types of membership operator</h4>

1. in

2. not in

<h4>1. In </h4>

Returns true if value exist in sequence

```python
num = [1,2,3,4,5]

print (3 in num)
```

```python
str = "Kajal"

print( "a" in str)
```

<h4>2. Not in </h4>

Returns true if value does not exist in sequence

```python
a = (1,2,3,4,5)

print(7 not in a)
```

<h3> Identity Operator</h3>

Used to check whether the two variables referes the same object in memory 

<h4> Types of Identity operator</h4>

1. Is

2. Is Not

<h4>1. Is</h4>

Returns the true if both variables refers same object

```python
a = [10,20,30,40,50]

b = a

print(a is b)
```

<h4>2.Is Not</h4>

Returns true if both variable refers two different objects

```python
S = [1,3,4,5,6]

T = [1,3,4,5,6]

print( S is not T)
```
