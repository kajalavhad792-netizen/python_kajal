<h4>Parameteres</h4>




<h4>1.Sep parameter</h4>

It specifies separator between multiple values by defualt it is space (" ")

<h4>Examples</h4>

```python
print("python","java","c++",sep = ", ")
```

```python
print(10,20,30,40,sep = ":")
```

```python
print(1,2,3,4,5,sep = ".")
```

```python
print("a","b","c","d",sep = "-")
```

```python
print("Start", "Middle", "End", sep=" -> ")
```

<h4>End parameter</h4>

It specifies what should printed at end of the output default it is a new line  ("\n")

```python
print("Hello", end=" ")
print("World")
```

```python
print("python", end=" ")
print("programming")
```

```python
print(12, end=" ")
print(21)
```

```python
print("kajal",end=" ")
print(11)
```

```python
print("kajal",end=",")
print(11)
```
