<h1>Data Type</h1>:-
In python defines the type of data a variable can store such as int, float, string, list, tuple, set, dictionary, boolean, and NoneType.

<h3>1.Integer</h3>:- Stores the whole numbers

Ex:- a = 10

<h3>2.Float</h3>
:- Stores numbers with decimal points

Ex:- a = 1.2

<h3>3.String</h3>
:-Stores the characters

Ex:- a = "kajal"

<h3>4.Boolean</h3>
:-Stores the logical values True and False  :: True:- 0  and False:- 1

Ex:- is_student = True

<h3>5.Complex</h3>
:- Stores a complex number having real and imaginary parts

Ex:- a = 2 + 3j

<h3>6.List</h3>
:- List a collection of elements
. list is mutable
. stores a multiple data in single variable
. list can store the different data types such as int,float,bool,complex 
. allows duplicate value

```python
lst = [10,20,30]
print (type(lst))
```

<h3>7.Tuple</h3>
:- Stores a multiple items in single variable

:- It represent using ()

:- It is a imutable means one we cretaed a tuple we cannot insert,modify,remove element

:- It allows duplicate values

:- Faster than list because its immutable

```python
student = (1,2,3,4,5)
print(student)
```

<h3>8.Set</h3


:- Set is a unoreder and mutable collection of unique elements

:- It represented using {}

:- It is a mutable means we can add or remove element after set creation

:- Duplicates values are not allowed

:- It can store different data types

```python
my_set = {10,20,30,40,50}
print(my_set)
```

<h3>9.Dictionary</h3>

:- in dictionary data is stord in key value-pair

:- It represents using { }

:- It is mutable means elements can add,remove,update

:- It stores different data types

:- Duplicate values cannot be allowed

```python
student = {
    "id":101,
     "name":"kajal",
      "age":21,
}
print(student)
```

<h3>Conversion Example</h3>

<h3>List to Tuple</h3>

```python
my_list = [10, 20, 30]

my_tuple = tuple(my_list)

print(my_tuple)
```

<h3>Tuple to list</h3>

```python
my_tuple = (10, 20, 30)

my_list = list(my_tuple)

print(my_list)
```

<h3>List to Set</h3>

```python
my_list = [10, 20, 30, 20]

my_set = set(my_list)

print(my_set)
```

<h3>Set to List</h3>

```python
my_set = {10, 20, 30}

my_list = list(my_set)

print(my_list)
```

<h3>Tuple to Set</h3>

```python
my_tuple = (10, 20, 30)

my_set = set(my_tuple)

print(my_set)
```

<h3>Set to Tuple</h3>

```python
my_set = {10, 20, 30}

my_tuple = tuple(my_set)

print(my_tuple)
```

<h3>String to List</h3>

```python
text = "Python"

my_list = list(text)

print(my_list)
```

<h3>List to String</h3>

```python
my_list = ['P', 'y', 't', 'h', 'o', 'n']

text = "".join(my_list)

print(text)
```

<h3>Dictionary to List</h3>

```python
my_dict = {"a": 1, "b": 2, "c": 3}

my_list = list(my_dict)

print(my_list)
```

<h3>List to Dictionary</h3>

```python
my_list = [('a', 1), ('b', 2), ('c', 3)]

my_dict = dict(my_list)

print(my_dict)
```
