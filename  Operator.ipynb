<h3>2.Comparison Operator</h3>

comparison operator compares two value

This always return boolean values True or False

They are also called as Relational operator

They are used for decision making


Types of Comparison Operator

1. Equal(=)

2. Not equal(!=)

3. Greater than(>)

4. Less than(<)

5. Greater than equals to(>=)

6. Less than equals to(<=)

```python
a = 10
b = 20

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

<h3>Logical Opeartor</h3>

1. Logical operator combines multiple conditions

2. They works with boolean values

3. And returns True only if all condition is True

4. Or returns True at least one condition is True

5. Not returns boolean values

```python
a = 10

b = 20

print(a < b and b > 15)
print(a > b and b > 15)
```

```python
x = 50
y = 30

print(x > y and y < 40)
print(x < y and y < 40)
```

<h3>OR</h3>

```python
a = 10

b = 20

print(a > b or b > 15)
print(a < b or b < 15)
```

```python
a = 10.5

b = -10

print(a > b or b > 15)
print(a < b or b < 15)
```

<H3>NOT</H3>

```python
a = 10

print(not(a > 5))
print(not(a < 10))
```
