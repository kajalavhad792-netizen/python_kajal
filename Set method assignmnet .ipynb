```python
# Question 6: Remove Banned Words 
'''
sentence = "python is easy and python is powerful" 
banned = {"is","and"} 
Print the sentence without banned words
'''

sentence = "python is easy and python is powerful" 
banned = {"is","and"}
words = sentence.split()
for word in words:
    if word not in banned:
        print(word,end=" ")
```

```python
# DICTIONARY ASSIGNMENTS (15 Questions) 
'''
Level 1 
Question 1: Student Marks 
students = { 
"Rahul":85, 
"Amit":72, 
"Sneha":94, 
"Pooja":67 
} 
Print only students scoring above 80.
'''

#Print only students scoring above 80.

students = { "Rahul":85, "Amit":72, "Sneha":94, "Pooja":67 }
for name, marks in students.items():
    if marks > 80:
        print(name, marks)
```

```python
# Question 2 
'''
Find 
● Highest Marks  
● Lowest Marks  
● Average Marks 
'''

students = { "Rahul":85, "Amit":72, "Sneha":94, "Pooja":67 }

highest = max(students.values())
lowest = min(students.values())
average = sum(students.values()) / len(students)
print("Highest Marks:", highest)
print("Lowest Marks:", lowest)
print("Average Marks:", average)
```

```python
# Question 3 
'''
Count frequency of characters. 
Input 
"programming" 
Expected 
{ 
'p':1, 
'r':2, 
... 
} 
'''

input = "programming"
frequency = {}
for char in input:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
print(frequency)
```

```python
# Question 4 
'''
Count word frequency. 
Input 
"python is easy python is powerful" 
'''

word = "python is easy python is powerful" 
frequency = {}
words = word.split()
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
print(frequency)        
```
