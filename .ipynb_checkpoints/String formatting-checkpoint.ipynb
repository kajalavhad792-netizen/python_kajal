<h3>String formatting</h3>

String formatting is the process of inserting variables, values, or expressions into a string to create a formatted output.

<h4>Types of string formatting</h4>

1. Concatenation  ( + )
2. Comma separation ( , )
3. Percentile formatting ( % )
4. . Format ( . )
5. f-strings

<h4>1. Concatenation</h4>

String Concatenation is the process of joining two or more strings together using the + operator.

<h4>Example</h4>

```python
first_name = "Kajal"
last_name = "Avhad"

print(first_name + " " + last_name)
```

```python
city = "Pune"
state = "Maharashtra"

print(city + ", " + state)
```

```python
name = "kajal"
clg  = "svit"

print(name +" "+ clg)
```

<h4>2.Comma separator</h4>

A comma separator is used in the print() function to print multiple values. By default, Python separates them with a single space.

```python
name = "kajal"
age = 21

print(name,age)
```

```python
name = "kajal"
age = 21
clg = "svit"

print(name,age,clg)
```

<h4>3. % formatting</h4>

% formatting is an old-style string formatting method used to insert values into a string using the % operator.

<h4>Common format specifiers</h4>

1. %s  =>       string                         => "kajal"
 
2. %d  =>       int                             => 21
 
3. %f  =>       float                           => 90.32
 
4. %.2f =>    Float (2 decimal places)          => 199.21
 
5. %c  =>      single character                =>'a'

<h4>1. %s </h4>

%s is a string format specifier used with % formatting. It inserts a string value into a formatted string.

```python
name = "Kajal"

print("My name is %s." % name)
```

```python
a = "good morning"

print("hi %s." % a)
```

```python
b = "everyone"

print("how are you %s." %b)
```

```python
fruit = "mango"

print("my favourite fruit is %s." %fruit)
```

<h4>2. %d</h4>

%d is an integer format specifier used with % formatting. It inserts an integer (whole number) into a formatted string.

```python
age = 22

print("Age: %d" % age)
```

```python
marks = 95

print("Marks: %d" % marks)
```

```python
id = 101

print("id: %d" %id)
```

<h4>3. %f</h4>

%f is a floating-point format specifier used with % formatting. It inserts a float (decimal) value into a formatted string.

```python
price = 99.99

print("Price: %f" % price)
```

```python
marks = 12.11

print("marks: %f" % marks)
```

<h4>4. %.2f</h4>

%.2f is a floating-point format specifier used with % formatting. It displays a float value with exactly 2 digits after the decimal point.


```python
price = 99.999

print("%.2f" % price)
```

```python
marks = 95.756

print("Marks = %.2f" % marks)
```

<h4>5. %c</h4>

%c is a character format specifier used with % formatting. It inserts a single character into a formatted string.


A single-character string (e.g., 'A')

```python
letter = 'A'

print("Letter: %c" % letter)
```

```python
grade = 'B'

print("Grade: %c" % grade)
```
