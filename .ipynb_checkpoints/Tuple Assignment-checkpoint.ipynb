<h3>Basic</h3>

1. Create two tuples and concatenate them.  
2. Repeat a tuple four times using the * operator.  
3. Check whether "Python" exists in a tuple.  
4. Print all tuple elements using a for loop.  
5. Print all tuple elements using a while loop. 

```python
# Create two tuples and concatenate them.

tuple1 = ("Apple", "Banana", "Mango")
tuple2 = ("Orange", "Grapes", "Pineapple")
print(tuple1 + tuple2)
```

```python
# Repeat a tuple four times using the * operator.

colors = ("Red", "Green", "Blue")
result = colors * 4
print(result)
```

```python
# Check whether "Python" exists in a tuple.

languages = ("Python", "Java", "C", "C++")
if "Python" in languages:
    print("Python is present")
else:
    print("Python is not present")
```

```python
# Print all tuple elements using a for loop.

fruits = ("Apple", "Banana", "Mango", "Orange")
for fruit in fruits:
    print(fruit,end=" ")
```

```python
# Print all tuple elements using a while loop.

fruits = ("Apple", "Banana", "Mango", "Orange")
i = 0
while i < len(fruits):
    print(fruits[i])
    i += 1
```

<h3>Intermediate</h3>

6. Find the maximum, minimum, total, and average of a tuple containing student marks.  
7. Create a tuple of product prices and calculate the total bill using sum().  
8. Check whether a user's role exists in a tuple of valid roles.  

```python
# Find the maximum, minimum, total, and average of a tuple containing student marks.

marks = (85, 92, 78, 88, 95)

maximum = max(marks)
minimum = min(marks)
total = sum(marks)
average = total / len(marks)

print("Maximum Marks:", maximum)
print("Minimum Marks:", minimum)
print("Total Marks:", total)
print("Average Marks:", average)
```

```python
# Create a tuple of product prices and calculate the total bill using sum().

prices = (250, 450, 120, 300, 180)
bill = sum(prices)
print("Bill:", bill)
```

```python
# Check whether a user's role exists in a tuple of valid roles.

roles = ("Admin", "Manager", "Employee", "Guest")
user_role = input("Enter your role: ")
if user_role in roles:
    print("Access Granted")
else:
    print("Access Denied")
```

<h3>Basic</h3>

1.Create a tuple using packing.  
2. Unpack a tuple into three variables.  
3. Swap two variables without using a temporary variable.  
4. Create a tuple of five numbers and unpack the first, middle, and last values using *. 

```python
# Create a tuple using packing.

student = "Rahul", 20, 85, "Nashik"
print(student)
```

```python
# Unpack a tuple into three variables

colors = ("Red", "Green", "Blue")
c1, c2, c3 = colors
print(c1)
print(c2)
print(c3)
```

```python
# Swap two variables without using a temporary variable.

a = 10
b = 20
a, b = b, a
print("a =", a)
print("b =", b)
```

```python
#  Create a tuple of five numbers and unpack the first, middle, and last values using *.

numbers = (10, 20, 30, 40, 50)
first, *middle, last = numbers
print("First :", first)
print("Middle:", middle)
print("Last  :", last)
```

```python
# Create a tuple of fruits and count how many times "Apple" appears.

fruits = ("Apple", "Mango", "Banana", "Apple", "Orange", "Apple")
count = fruits.count("Apple")
```

```python
#  Find the index of "Mango" in a tuple.
fruits = ("Apple", "Mango", "Banana", "Apple", "Orange", "Apple")
l=fruits.index("Mango")
print(l)
```

```python
# Create a tuple of numbers and count the occurrence of 5.
list = [1,2,3,4,5]
tuple=tuple(list)
print(tuple)
```

```python
# Create a tuple of numbers and count the occurrence of 5.
numbers = (5, 2, 8, 5, 10, 5, 12)
count = numbers.count(5)
print(count)
```

```python
# Create a tuple of employee names and check whether a given employee exists before finding the index.

emp = ("kajal", "tejal", "rajashri", "arshil")
name = input("Enter employee name: ")
if name in emp:
    print("Employee found at index:", emp.index(name))
else:
    print("Employee does not exist.")
```

```python
# Compare a list and a tuple using sys.getsizeof() to observe memory usage. 

import sys

my_list = [10, 20, 30, 40, 50]
my_tuple = (10, 20, 30, 40, 50)
print("List memory:", sys.getsizeof(my_list), "bytes")
print("Tuple memory:", sys.getsizeof(my_tuple), "bytes")
```

```python
# Create a tuple of months and print the index of "July". 

months = (
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
)
print("Index of July:", months.index("July"))
```
