<h3>2.Comparison Operator</h3>

1. Comparison opeartor compare two values
2. They are also called Relational operator
3. They also returns boolean values (True or False)
4. Comparison operatores compare numbers, string, and other objects

<h4>Types of Comparison Operator</h4>

1. Equals to (=)
2. Not equals to (!=)
3. Greater than (>)
4. Less than (<)
5. Greater than or equals to (>=)
6. Less than or equals to (<=)

<h4>Examples</h4>

```python
a = 10

b = 20

print(a == b)   
print(a != b)   
print(a > b) 
print(a < b)    
print(a >= b)   
print(a <= b) 
```

```python
c = 10.5

d = 20.1

print(c == d)
print(c != d)
print(c > d)
print(c < d)
print(c >= d)
print(c <= d)
```

```python
a = "Apple"
b = "Banana"

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
a = 15
b = 10

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

```python
a = 25
b = 15

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

<h3>3.Logical Operator</h3>

1. Logical operators combine multiple conditions
2. Python provids three logical operators
3. AND = all conditions must be true
4. NOT = reverse the boolean values
5. OR = one condition must be true

<h4>Examples</h4>

1. <h5>AND</h5>

```python
a = 10
b = 20

print(a < b and b > 15)
```

```python
a = 30
b = 15

print(a < b and b > 10)
```

<h5>NOT</h5>

```python
a = 10

print(not(a > 5))
```

```python
a = 3

print(not(a > 5))
```

<h5>OR</h5>

```python
a = 10
b = 20

print(a > b or b > 15)
```

```python
a = 5
b = 10

print(a > b or b < 5)
```
